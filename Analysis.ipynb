## Missing Value Treatment

### Remove the records where Closed Date values are null

```python
df[df['Closed Date'].isnull() == True]
```

```python
df = df[df['Closed Date'].notna()]
```

```python
df.shape
```

## Analyze the date column and remove entries that have an incorrect timeline

```python
# time taken to solve the case
df['time taken'] = df['Closed Date'] - df['Created Date']
```

```python
# convert to seconds
df['time taken'] = df['time taken'].astype(np.int64)
```

```python
df['time taken']
```

```python
df[df['time taken'] > 0]
```

```python
df['time taken'].describe()
```

### frequency plot for complaints in each city

```python
# count the number of complaints in each city
complaints_citywise = df.groupby('City')['Complaint Type'].count()
```

```python
complaints_citywise = complaints_citywise.sort_values(by='')
```

```python
result_df = pd.DataFrame(complaints_citywise).reset_index()
```

```python
result_df = result_df.sort_values(by='Complaint Type', ascending=False)
```

```python
sns.catplot(data=result_df.head(10), kind='bar',  x='Complaint Type', y='City')
plt.ylabel('Number of Complaints')
plt.show()
```
