## Bivariate Analysis

```python
sns.countplot(x="Pclass", hue="Survived", data=train_df)
```

```python
#more number of people from Pclass '1' have survived
```

```python
sns.countplot(x="Embarked", hue="Survived", data=train_df)
```

```python
#more number of people are likely to survive with Embarked 'Q'
```

```python
sns.countplot(x="SibSp", hue="Survived", data=train_df)
```

```python
sns.countplot(x="Sex", hue="Survived", data=train_df)
```

```python
#more number of females have survive
```

```python
sns.catplot(x="Sex", y="Fare", hue="Survived", data=train_df, kind="bar")
sns.catplot(x="Sex", y="Survived", hue="Embarked", data=train_df, kind="bar")
sns.catplot(x="Sex", y="Survived", hue="Pclass", data=train_df, kind="bar")
```

```python
"""
OBSERVATIONS
1. The people with high fare have survived than people with low fair
2. Most females that have survived have port of embarkation as Cherbourg
3. Females having Pclass as 1 & 2 have survived more than Pclass 3
"""
```

```python
sns.catplot(x="Parch", y="Survived", hue="Pclass", data=train_df, kind="bar")
```

```python
sns.catplot(x="SibSp", y="Survived", hue="Pclass", data=train_df, kind="bar")
```
